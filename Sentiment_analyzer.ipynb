```python
from sklearn.datasets import fetch_20newsgroups
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
import nltk
import re
```

```python
text_data=fetch_20newsgroups()
```

```python
raw_text=text_data.data[:4]
raw_text
```

Convert to Lower case

```python
lower_text=[]
for i in raw_text:
    lower_text.append(i.lower())
```

```python
lower_text
```

Tokenize the sentences to words

```python
tokenized_text=[ word_tokenize(i) for i in lower_text]
```

Filter by removing symbols punctuations etc.

```python
filtered_text=[]
for i in range(0,len(tokenized_text)):
    filter_sent=[]
    for word in tokenized_text[i]:
        res=re.sub(r'[^\w\s]','',word)
        if(res!=''):
            filter_sent.append(res)
    filtered_text.append(filter_sent)
        
```

```python
filtered_text
```

#Stopword Removal

```python
stopword_filtered=[]
for i in range(0,len(filtered_text)):
    temp=[]
    for word in filtered_text[i]:
        if word not in stopwords.words('english'):
            temp.append(word)
    stopword_filtered.append(temp)
stopword_filtered
```

#Stemming using porter stemmer

```python
port_stemmer=PorterStemmer()
stemmed_words=[]
for i in range(0,len(stopword_filtered)):
    temp=[]
    for word in stopword_filtered[i]:
        temp.append(port_stemmer.stem(word))
    stemmed_words.append(temp)
    
        
```

```python
stemmed_words
```

#LemmatizedWords

```python
w_net=WordNetLemmatizer()
nltk.download('wordnet')
```

```python
lemmatized_words=[]
for i in range(0,len(stemmed_words)):
    temp=[]
    for word in stemmed_words[i]:
        temp.append(w_net.lemmatize(word))
    lemmatized_words.append(temp)
lemmatized_words
```
